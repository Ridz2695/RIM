##### Creating a list of all unique schools across SY 18 and SY 19

```python
schoolNamesfy18=fy18['School_FullName'].tolist()
schoolNamesfy19=fy19['School_FullName'].tolist()
print (len(list(set(schoolNamesfy18))))
print (len(list(set(schoolNamesfy19))))
```

```python
schoolNames = set(schoolNamesfy18 + schoolNamesfy19)
#print (schoolNames)
schoolN=list(schoolNames)
#print (schoolN)
school_table=pd.DataFrame(schoolN, columns=['SchoolName'])
school_table=school_table.dropna()
school_table=school_table.sort_values(['SchoolName'])
print(len(school_table['SchoolName']))
```

##### There are 50 unique schools across SY 18 and SY 19
##### The following chunk of code will assign a school id to each school

```python
school_table['school_id'] = (school_table.groupby(['SchoolName']).cumcount()==0).astype(int)
school_table['school_id'] = school_table['school_id'].cumsum()
```

```python
school_table['initials']=np.where(school_table['school_id']<10, 'sc10','sc1')
school_table['sch_id']= school_table['initials'] + school_table.school_id.map(str)
header = ["SchoolName", "sch_id"]
school_table.to_csv('/Users/snigd/Desktop/Dropbox/UIC Data Project/Working/tables/school_table.csv', columns = header, index=False)
```
