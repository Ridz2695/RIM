```python
# import libraries 
import pandas as pd 
import numpy as np
```

```python
# data loading 
nonstudent_table=pd.read_csv("/Users/fudanyi/Dropbox/UIC Data Project/Working/tables/coach_table.csv")
fy18_original=pd.read_csv("/Users/fudanyi/Documents/2019 Fall Term/560 Capston/data_processing_code/original_data/fy18.csv")
```

```python
nonstudent_table.rename(columns = {'_merge':'merge','SchoolName':'School_FullName',"TeacherName":"Teacher","CoachName":"Coach"}, inplace = True)
```

```python
nonstudent_table.head(5)
```

```python
#df.loc[df['column_name'] == some_value]
nonstudent_table_for_18=nonstudent_table.loc[nonstudent_table["merge"]!="sy19"]
#nonstudent_table_2["merge"].value_counts()
```

```python
fy18_original.head(5)
#fy18_original.columns
```

### Do the mapping for school and school_id

```python
school_dic=dict(zip(nonstudent_table_for_18.School_FullName,nonstudent_table_for_18.sch_id))
#school_dic
```

```python
#fy18_original["sch_id"]=""
```

```python
#df['col1'].map(di) 
#replace(di, inplace=True)
#data['classes_wild'] = data['category_id'].apply(lambda cw: classes_wild[cw])

#fy18_original["sch_id"]=replace(school_dic, inplace=True)
```

```python
fy18_original["sch_id"]=fy18_original["School_FullName"].apply(lambda sid: school_dic[sid])
```

### Do the mapping for teacher and teacher_id

```python
teac_dic=dict(zip(nonstudent_table_for_18.TeacherName,nonstudent_table_for_18.tch_id))
```

```python
fy18_original["tch_id"]=fy18_original["Teacher"].apply(lambda tid: teac_dic[tid])
```

### Do the mapping for coach and coach_id

```python
coach_dic=dict(zip(nonstudent_table_for_18.Coach,nonstudent_table_for_18.coach_id))
```

```python
fy18_original["coach_id"]=fy18_original["Coach"].apply(lambda cid: coach_dic[cid])
```
