```python
allfy1819.fillna('',inplace=True)
allfy1819
```

```python
allfy1819['initials']=np.where(allfy1819['student_id']<10, 'st000','st')
allfy1819['initials']=np.where(allfy1819['student_id']>=10, 'st00',allfy1819['initials'])
allfy1819['initials']=np.where(allfy1819['student_id']>=100, 'st0',allfy1819['initials'])
allfy1819['initials']=np.where(allfy1819['student_id']>=1000, 'st',allfy1819['initials'])
allfy1819['student_id']= allfy1819['initials'] + allfy1819.student_id.map(str)
allfy1819
```

```python
allfy1819=allfy1819.sort_values(['sch_id','student_id'])
allfy1819
```
