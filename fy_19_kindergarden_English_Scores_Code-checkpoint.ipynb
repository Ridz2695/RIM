## fill in scores for fsf

```python
fsf=fy19e[fy19e['TestTypeCode']=='FSF']
```

```python
fsf=fsf[['student_id','BOY','SEP','OCT','NOV','DEC','MOY']]
```

```python
fsf[fsf.duplicated(['student_id'], keep=False)].sort_values(['student_id'])
```

```python
fsf['MOY']=fsf.groupby(['student_id'], sort=False)['MOY'].apply(lambda x: x.bfill())
```

```python
fsf['DEC']=fsf.groupby(['student_id'], sort=False)['DEC'].apply(lambda x: x.bfill())
```

```python
fsf['NOV']=fsf.groupby(['student_id'], sort=False)['NOV'].apply(lambda x: x.bfill())
```

```python
fsf['OCT']=fsf.groupby(['student_id'], sort=False)['OCT'].apply(lambda x: x.bfill())
```

```python
fsf['SEP']=fsf.groupby(['student_id'], sort=False)['SEP'].apply(lambda x: x.bfill())
```

```python
fsf['BOY']=fsf.groupby(['student_id'], sort=False)['BOY'].apply(lambda x: x.bfill())
```

```python
fsf[fsf.duplicated(['student_id'], keep=False)].sort_values(['student_id'])
```

```python
fsf.rename({'MOY':'MOY_FSF'},axis='columns' ,inplace=True)
```

```python
fsf['fpm_fsf']=fsf['OCT']
fsf['fpm_fsf']=np.where((pd.isnull(fsf.fpm_fsf))&(fsf['NOV']!=np.NaN),fsf['NOV'],fsf['fpm_fsf'])
fsf['fpm_fsf']=np.where((pd.isnull(fsf.fpm_fsf))&(fsf['DEC']!=np.NaN),fsf['NOV'],fsf['fpm_fsf'])
fsf['fpm_fsf']=np.where((pd.isnull(fsf.fpm_fsf))&(fsf['SEP']!=np.NaN),fsf['NOV'],fsf['fpm_fsf'])
```

```python
fsf.drop(['SEP','OCT','NOV','DEC'], axis=1, inplace=True)
```

# fill in scores for NWF-CLS

```python
nwf=fy19e[fy19e['TestTypeCode']=='NWF-CLS']
```

```python
nwf=nwf[['student_id','MOY','JAN','FEB','Mar','APR','EOY']]
```

```python
nwf[nwf.duplicated(['student_id'], keep=False)].sort_values(['student_id'])
```

```python
nwf['MOY']=nwf.groupby(['student_id'], sort=False)['MOY'].apply(lambda x: x.bfill())
nwf['JAN']=nwf.groupby(['student_id'], sort=False)['JAN'].apply(lambda x: x.bfill())
nwf['FEB']=nwf.groupby(['student_id'], sort=False)['FEB'].apply(lambda x: x.bfill())
nwf['Mar']=nwf.groupby(['student_id'], sort=False)['Mar'].apply(lambda x: x.bfill())
nwf['APR']=nwf.groupby(['student_id'], sort=False)['APR'].apply(lambda x: x.bfill())
nwf['EOY']=nwf.groupby(['student_id'], sort=False)['EOY'].apply(lambda x: x.bfill())
```

```python
nwf.drop_duplicates(subset=['student_id'],keep='first', inplace=True)
```

```python
nwf.rename({'MOY':'MOY_NWF','EOY':'EOY_NWF'},axis='columns' ,inplace=True)
```

```python
nwf['spm_nwf']=nwf['Mar']
nwf['spm_nwf']=np.where((pd.isnull(nwf.spm_nwf))&(nwf['APR']!=np.NaN),nwf['APR'],nwf['spm_nwf'])
nwf['spm_nwf']=np.where((pd.isnull(nwf.spm_nwf))&(nwf['JAN']!=np.NaN),nwf['APR'],nwf['spm_nwf'])
nwf['spm_nwf']=np.where((pd.isnull(nwf.spm_nwf))&(nwf['FEB']!=np.NaN),nwf['APR'],nwf['spm_nwf'])
```

```python
nwf.drop(['JAN','FEB','Mar','APR'], axis=1, inplace=True)
```
